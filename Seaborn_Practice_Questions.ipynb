## Question 1: Comparing Two Lists Using a Line Plot

Problem Statement:
You are given two lists representing monthly sales of two products over 6 months.
You need to visualize how their sales trends compare over time using a line plot.

Approach:
1. Create two lists: months and sales.
2. Use `sns.lineplot()` to plot both lines on the same graph.
3. Add appropriate labels and a title to make the visualization clear.

```python
# Write your code here
import seaborn as sns
import matplotlib.pyplot as plt

# Example lists
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales_A = [45, 55, 60, 70, 65, 75]
sales_B = [40, 50, 58, 68, 70, 80]

# Your line plot code here
a=sns.lineplot(x=months,y=sales_A,label='Product A')
b=sns.lineplot(x=months,y=sales_B,label='Product B')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales Comparison')
plt.legend()
```

## Question 10: Average Tip by Distance Category

Problem Statement:
We want to know if longer trips generally result in higher tips.

Approach:
1. Create distance categories (e.g., short, medium, long) using the 'distance' column.
2. Use `sns.barplot()` to plot average tips for each distance category.

```python
# Write your code here
taxis = sns.load_dataset('taxis')
taxis['distance_category'] = pd.cut(taxis['distance'], bins=[0, 5, 10, float('inf')], labels=['short', 'medium', 'long'])
sns.barplot(x='distance_category', y='tip', data=taxis)
plt.xlabel('Distance Category')
plt.ylabel('Average Tip')
plt.title('Average Tip by Distance Category')
plt.show()
```
